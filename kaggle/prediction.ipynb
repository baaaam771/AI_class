```python
def process_json_data(series):
    str_list = []
    for i in series:
        string = []
        if (i != '0'):
            o = ast.literal_eval(i)
            for i in o:
                for j in i.items():
                    if (j[0] == 'name'):
                        string.append(j[1])
        str_list.append(' + '.join(string))
    return LabelEncoder().fit_transform(str_list)
```
