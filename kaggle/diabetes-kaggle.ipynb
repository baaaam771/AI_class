```python
import pandas as pd
```

```python
train_data = pd.read_csv('../input/diabetes-classification/train.csv')
test_data = pd.read_csv('../input/diabetes-classification/test.csv')
```

```python
train_data.head()
```

```python
from keras import regularizers
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
# preprocessing.MinMaxScaler
# 스케일 된 값을 별도의 변수로 저장
# pd.DataFrame 로 train_data
min_max_scaler = preprocessing.MinMaxScaler()
x = train_data.values
x_scaled = min_max_scaler.fit_transform(x)
train_data = pd.DataFrame(x_scaled)
```

```python
train_data.head()
```

```python
y = test_data.values
y_scaled = min_max_scaler.fit_transform(y)
test_data = pd.DataFrame(y_scaled)
```

```python
# train, valid train_test_split // 8 : 2
train_data, valid_data = train_test_split(train_data, test_size=0.2)
```

```python
train_data.shape, valid_data.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
# 모델 구성
# DNN
# 이진 분류

model = Sequential([
    Dense(200, input_dim=8, activation='relu'),
    Dense(100, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
adam = optimizers.Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
```

```python
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
train_x = train_data.iloc[:,1:9]
train_y = train_data.iloc[:,9]
```

```python
# fit
history = model.fit(train_x, train_y, epochs=100, batch_size=10)
```

```python
val_loss, val_acc = model.evaluate(valid_data.iloc[:,1:9], valid_data.iloc[:,9])
```

```python
test_x = test_data.iloc[:,1:9]
```

```python
results = model.predict(test_x)
```

```python
results
```

```python
rounded = [int(round(x[0])) for x in results]
```

```python
rounded
```
