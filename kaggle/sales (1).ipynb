```python
import numpy as np
import pandas as pd
import os
```

```python
# read all csv

sales_data = pd.read_csv('../input/competitive-data-science-predict-future-sales/sales_train.csv')
item_cat = pd.read_csv('../input/competitive-data-science-predict-future-sales/item_categories.csv')
items = pd.read_csv('../input/competitive-data-science-predict-future-sales/items.csv')
shops = pd.read_csv('../input/competitive-data-science-predict-future-sales/shops.csv')
test_data = pd.read_csv('../input/competitive-data-science-predict-future-sales/test.csv')
```

```python
sales_data.head()
```

```python
dataset = sales_data.pivot_table(index=['shop_id', 'item_id'], 
                                 values=['item_cnt_day'], 
                                 columns=['date_block_num'], 
                                 fill_value=0, 
                                 aggfunc='sum'
                                )
```

```python
dataset.reset_index(inplace=True)
```

```python
dataset.head()
```

```python
dataset = pd.merge(test_data, dataset, on=['item_id', 'shop_id'], how='left')
```

```python
dataset.fillna(0, inplace=True)
```

```python
dataset.head()
```

```python
dataset.drop(['ID', 'shop_id', 'item_id'], inplace=True, axis=1)
dataset.head()
```

```python
x_train = np.expand_dims(dataset.values[:,:-1], axis=2)
y_train = dataset.values[:,-1:]
```

```python
x_test = np.expand_dims(dataset.values[:,1:], axis=2)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential([
    LSTM(64, input_shape=(33,1)),
    Dropout(0.4),
    Dense(1)
])
```

```python
model.compile(loss='mse', metrics=['mae'])
```

```python
model.fit(x_train, y_train, batch_size=512, epochs=10)
```

```python
predictions = model.predict(x_test)
```

```python
submission = pd.DataFrame({'ID' : test_data['ID'], 'item_cnt_month' : predictions.ravel()})
```

```python
submission.to_csv('pred.csv', index=False)
```
